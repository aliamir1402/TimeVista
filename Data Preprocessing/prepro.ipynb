```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Rawalakot_data_1980_2022.csv")
data['date'] = pd.to_datetime(data['date'])
# data=pd.read_csv("Hourly_data.csv")
data.head()
```

```python
for col in data.columns[1:]:
# for col in data.columns:
    plt.figure()
    
    # boxplots
    plt.boxplot(data[col])
    plt.title(str(col))
    plt.show()

    # histograms
    plt.hist(data[col])
    plt.title(str(col))
    plt.show()
    
    # kernel density plots
    sns.kdeplot(data[col])
    plt.title(str(col))
    plt.show()
    

# print(type(list(data.columns)))
# print(list(data.columns))

# plt.boxplot(data['dewpoint_temperature_2m_mean'])
# plt.title("dewpoint_temperature_2m_mean")
# plt.show()
```
